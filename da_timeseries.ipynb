 ## Resampling

lower frequency (downsampling) thng lin quan ti hot ng tng hp, v d mc doanh thu
trong thng t d liu ngy. Resampling cho higher grequency (upsampling) t ph bin hn,

```python
data_columns = ['Consumption', 'Wind', 'Solar','Wind+Solar']
# Resample from day to weekly frequency, aggretgating with mean 
opsd_weekly_mean = data[data_columns].resample('W').mean()
opsd_weekly_mean.head(3)
```

```python
start,end ='2017-01','2017-06'
fig,ax = plt.subplots()
ax.plot(data.loc[start:end,'Solar'], marker = '.',linestyle = '-', linewidth = 0.5,label = 'Daily' )
ax.plot(opsd_weekly_mean.loc[start:end,'Solar'],marker = '.',linestyle = '-', linewidth = 0.5,label = 'Weekly Mean Resample')
ax.set_ylabel('Solar production (gwh)')
ax.legend()
plt.show()
```
